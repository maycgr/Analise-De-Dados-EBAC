## 1\. Preços `outliers` de diamante

```
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```
# gráfico do exercício 1

diamonds = data[['price']]
diamonds

with sns.axes_style('darkgrid'):
  grafico_1 = sns.boxplot(x='price',data=diamonds)
  grafico_1.set(title='Boxplot: outliers de preço dos diamantes',
                xlabel='Preço (U$D)')
```

**Insight do gráfico 1**: Os preços outliers de diamantes refletem uma grande variação nos valores. Outliers superiores podem representar diamantes raros ou de coleção, enquanto outliers inferiores podem resultar de imperfeições ou subvalorização. Esses valores extremos são influenciados por fatores externos, como demanda e especulação, oferecendo oportunidades para encontrar diamantes de alta qualidade com boa relação custo-benefício. Em suma, os outliers nos preços dos diamantes revelam a complexidade e a diversidade do mercado.

## 2\. Correlação entre o do preço do diamante com seu peso (`carat`) agrupados por sua transparêcia (`clarity`)

```
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```
# gráfico do exercício 2

diamonds = data[['price', 'carat', 'clarity']]
diamonds_grouped = diamonds.groupby('clarity').mean().corr()

with sns.axes_style('darkgrid'):
  grafico_2 = sns.heatmap(diamonds_grouped, annot=True)
```

**Insight do gráfico 2**: O heatmap mostra a relação entre o peso e o preço dos diamantes. Se houver uma correlação forte e positiva, diamantes mais pesados geralmente têm preços mais altos. No entanto, se a correlação for fraca ou negativa, outras características, como a clareza, podem ser mais importantes para determinar o preço. Isso ajuda a entender como diferentes atributos influenciam o valor dos diamantes no mercado.

## 3\. Distribuição contínua aproximada do peso (`weight`) de carros

```
import seaborn as sns

data = sns.load_dataset("mpg")
data.head()
```

```
# gráfico do exercício 3

cars = data[['weight']]

with sns.axes_style('darkgrid'):
  grafico_3 = sns.histplot(x='weight',data=cars,kde=True)
  grafico_3.set(title='Distribuição do peso de carros',
                xlabel='peso (u.m.)',
                ylabel='Contagem')
```

**Insight do gráfico 3**: O histograma mostra a distribuição contínua aproximada do peso dos carros. A concentração dos dados em uma faixa específica sugere um padrão na distribuição dos pesos dos carros.

## 1\. Preço do diamante por tipo de corte

````
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
````

````
diamouns = data[["cut", "price"]].groupby("cut").agg("sum").reset_index()
diamouns.head()
````

````
# gráfico do exercício 1
grafico = sns.barplot(data = diamouns , x="cut", y="price", ci=None, palette="pastel")
grafico.set(title='Preço do diamante por tipo de corte', xlabel='Tipo de corte', ylabel='Preço');
````

**Insight do gráfico 1**:O corte ideal tem o preço total mais alto, sugerindo que diamantes com esse tipo de corte são geralmente os mais caros devido à sua alta qualidade e precisão. Diamantes com corte premium têm o segundo preço total mais alto, indicando que eles também são altamente valorizados, embora possam ser um pouco menos caros do que os diamantes com corte ideal.

## 3\. Numero de passageiros por mês entre 1949 e 1959

````
import seaborn as sns

data_flights_2 = sns.load_dataset("flights")
data_flights_2.head()
````

````
# gráfico do exercício 3
flights_month = data_flights_2.query('year >= 1949 and year <= 1959')
with sns.axes_style('whitegrid'):
  grafico_flights_month = sns.lineplot(x='month', y='passengers', hue='year', data=flights_month);
  grafico_flights_month.set(title = 'Número de passageiros por mês', xlabel = 'Mês', ylabel = 'Passageiros')
  grafico_flights_month.grid(alpha=0.35)
  grafico_flights_month.legend(title='Ano',loc=[1.01,0])
````

**Insight do gráfico 3**: Dentro de cada ano, há variação mensal no número de passageiro com picos de passageiros durante os meses de verão e quedas durante os meses de inverno.
